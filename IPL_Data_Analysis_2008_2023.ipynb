```python
season_count = data['season'].value_counts().sort_index()
plt.figure(figsize=(10, 6))
season_count.plot(kind='bar', color='skyblue')
plt.title('Number of Matches Played in Each Season')
plt.xlabel('Season')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
toss_won = data['toss_won'].value_counts().sort_values(ascending=True)
toss_won
```

```python
toss_won.plot(kind='bar',color='#2a5233')
plt.title('Number of Tosses won by each Team')
plt.xlabel('Number of Tosses Won')
plt.ylabel('Team Name')
plt.show()
```

## team most won toss  from the season 2008 to 2023

```python
sns.boxplot(x='toss_won',y='season',data=data)
plt.xticks(rotation=45)
```

## Selection of BAT And Bowl based on Toss

```python
bowl_first = data['decision'].value_counts()[0]
bat_first = data['decision'].value_counts()[1]
plt.pie([bowl_first,bat_first],labels=['Bowl First','Bat First'],autopct='%1.1f%%')
plt.title('Bat First VS Bowl First Ratio')
plt.show()
```

## Percentage of Winning or Loosing Match After Winning Toss

```python
success = data[data['toss_won'] == data['winner']].shape[0]
failure = data[data['toss_won'] != data['winner']].shape[0]

total_matches = success + failure

winning_perc = round(success/total_matches*100,2)
failure_perc = round(failure/total_matches*100,2)

print('Winning perc:', winning_perc)
print('Loosing perc:', failure_perc)
```

```python
plt.bar(['Win','Loss'], [winning_perc,failure_perc],width=0.5)
plt.ylabel("Percentage (in %)")
plt.title("Percentage of Winning or Loosing Match After Winning Toss")
plt.show()
```

## stadium with most number of matches played

## Scores per Venue

```python
venue_avg_scores = data.groupby('venue_name')['home_runs'].mean()

sorted_venues = venue_avg_scores.sort_values(ascending=False)

plt.figure(figsize=(10, 6))
plt.bar(sorted_venues.index, sorted_venues.values, color='lightblue')
plt.title('Venue Analysis')
plt.xlabel('Venue')
plt.ylabel('Average Scores')
plt.xticks(rotation=90)
plt.show()
```

## Run Distribution

```python
runs = data['home_runs'].dropna().astype(int)

plt.figure(figsize=(8, 6))
plt.hist(runs, bins=20, color='orange', edgecolor='black')
plt.title('Run Distribution')
plt.xlabel('Runs')
plt.ylabel('Frequency')
plt.show()
```

## Match Win Percentage Heat Map

```python
win_percentage = data.groupby(['season', 'home_team']).apply(lambda x: (x['home_team'] == x['winner']).mean()).unstack()

plt.figure(figsize=(12, 8))
sns.heatmap(win_percentage, cmap='YlGnBu', annot=True, fmt='.2f', cbar=True)
plt.title('Match Win Percentage by Team and Season')
plt.xlabel('Home Team')
plt.ylabel('Season')
plt.show()
```

```python
win_percentage = data.groupby(['season', 'away_team']).apply(lambda x: (x['home_team'] == x['winner']).mean()).unstack()

plt.figure(figsize=(12, 8))
sns.heatmap(win_percentage, cmap='YlGnBu', annot=True, fmt='.2f', cbar=True)
plt.title('Match Win Percentage by Team and Season')
plt.xlabel('Home Team')
plt.ylabel('Season')
plt.show()
```
